### 3.4선형회귀 최소제곱법 실습

```
import numpy as np

x=[2,4,6,8]
y=[81,93,91,97]

mx = np.mean(x)
my = np.mean(y)


# 기울기 분모
divisor =0
for i in x:
  divisor += (i-mx)**2
print(divisor)
# 파이썬 반복문 이용 시그마 구현 위와 같음
divisor2 = sum([(i-mx)**2 for i in x])
print("divisor2=",divisor2)

#기울기 분자
divider=0
for i in range(len(x)):
  divider += (x[i]-mx)*(y[i]-my)

print(divider)

# 기울기
a= divider/divisor
print(a)

#절편
b=my - (mx*a)
print(b)


# 10시간 공부했을 때 예측 점수
y_pred = a*10+b

print("y=",y_pred)

```

### 3.7 평균제곱 오차

#### 내 풀이

```
#임이의 기울기와 절편
fake_a_b=[3,76]

# x,y 데이터
data = [[2,81],[4,93],[6,91],[8,97]]

x= [i[0] for i in data]
y = [i[1] for i in data]

#예측값
y_hat=[]
for i in x:
    y_hat.append(fake_a_b[0]*i + fake_a_b[1])
print(y_hat)

# 평균 제곱 오차
sum=0
for i in range(len(x)):
  sum+=( y[i]-y_hat[i])**2

mse = sum/len(x)
print(mse)
```
