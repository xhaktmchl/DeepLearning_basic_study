# 4장 경사하강법

### 경사하강법 실습

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


```

```
# 데이터 배열 x고웁시간 y 성적
data = [[2,81],[4,93],[6,91],[8,97]]
x = [i[0] for i in data]
y = [i[1] for i in data]

# 그래프 나타내기 
plt.figure(figsize=(8,5))
plt.scatter(x,y)
plt.show
```

```
#넘파이 배열생성
x_data = np.array(x)
y_data = np.array(y)

# 기울기,절편
a=b=0

# 학습률 설정
lr=0.03

epoch = 2001
# 경사 하강법 
for i in range(epoch):
  y_pred = a*x_data+b
  error = y_data - y_pred# 오차
  # ab 미분값
  a_diff = -2/len(x_data)*sum((x_data)*error)
  b_diff = -2/len(x_data)*sum(error)
  a = a-lr*a_diff
  b = b-lr*b_diff

  if i %100 == 0:
    print("에포크=%.4f, 기울기a=%.4f, 절편b=%f" %(i,a,b))

# 그래프 표현
plt.figure(figsize=(8,5))
plt.scatter(x,y)
plt.plot(x_data,y_pred)
plt.show()
    
  
```

주의 : 기울기 a 와 절편 b 각각 미분해서 경사하강법 이용
