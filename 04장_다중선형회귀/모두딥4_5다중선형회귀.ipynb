###4.5 다중선형회귀 실습

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```
# 데이터 공부시간,과외횟수, 점수

data = [[2,0,81],[4,4,93],[6,2,91],[8,3,97]]
x1 = [i[0] for i in data]
x2 = [i[1] for i in data]
y = [i[2] for i in data]

# 3차원 그래프 시각화
ax = plt.axes(projection='3d')
ax.set_xlabel('study_hours')
ax.set_ylabel('academy')
ax.set_zlabel('score')
ax.dist = 11
ax.scatter(x1,x2,y)
plt.show()

# 넘파이 배열로 변환
x1_data = np.array(x1)
x2_data = np.array(x2)
y_data = np.array(y)

# 경사하강법 
a1=a2=0
b=0
lr = 0.02 # 학습률
epoch = 2001
for i in range(epoch):
  y_pred= a1*x1_data + a2*x2_data + b
  error = y_data - y_pred
  a1_diff = -(2/len(x1))*sum(error*x1_data)
  a2_diff = -(2/len(x1))*sum(error*x2_data)
  b_diff = -(2/len(x1))*sum(error)

  a1 = a1-lr*a1_diff
  a2 = a2-lr*a2_diff
  b = b-lr*b_diff

  if(i % 100 == 0):
    print("epoch=%.2f 기울기1=%.2f 기울기2=%.2f 절편=%.2f"%(i,a1,a2,b))



```
