```
data =[[2,0],[4,0],[6,0],[8,1],[10,1],[12,1],[14,1]]

x = [i[0] for i in data] # 공부시간
y = [i[1] for i in data] # 합격여부

x_data = np.array(x)
y_data = np.array(y)

print(x_data*y_data)

print(x)

print(x_data)
print(x+y)
print(x_data+y_data)



# 데이터 시각화
plt.scatter(x,y) # 점으로 나타내기
plt.plot(x,y) # 선으로 나타내기
# x,y 범위 설정
plt.xlim(0,15) 
plt.ylim(-0.1,1.1)
plt.show()

# 시그모이드 함수
def sigmoid(x):
  return 1/(1+np.exp(-x))

# a,b 설정
a=b=0

# 학습률
lr = 0.05

# 경사하강법
epoch=2001
for i in range(epoch):
  a_diff= x_data*(sigmoid(x_data*a+b)-y_data)
  b_diff = sigmoid(a*x_data + b) - y_data

  a= a- lr*a_diff
  b= b- lr*b_diff
  y_pred = sigmoid(a*x_data+b)

  if i%100 == 0:
    print("에포크%.4f " %(i))
    print("a=",a)
    print("b=",b)

# 학습시킨 로지스틱 회귀 함수 시각화
#x_array= [i for i in range(0,15,0.1)]
x_array= np.arange(0,15,0.1)
plt.scatter(x_data,y_data)
plt.plot(x_array,np.array([sigmoid(a*x+b) for x in x_array]))

dd = [i for i in np.arange(0,15,0.1)]
print(dd)
x_da = np.array([i for i in np.arange(0,15,0.1)])
y_da = sigmoid(a*dd+b)
plt.plot(x_da,y_da)
  


```
